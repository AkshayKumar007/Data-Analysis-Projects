## 3. Data Cleaning

```python
# changing column labels
col_name = ['patient_id', 'appointment_id', 'gender', 'schedule_day', 'appointment_day', 'age', 'neighbourhood',
         'scholarship', 'hypertension', 'diabetes', 'alcoholism', 'handicap', 'sms_received', 'no_show']
df.columns = col_name
df.info()
```

```python
# change datatype of patient_id to int
df['patient_id'] = df['patient_id'].astype(int)
df.head()
```

```python
# convertong dates from string to datetime object
df['schedule_day'] = pd.to_datetime(df['schedule_day'])
df['appointment_day'] = pd.to_datetime(df['appointment_day'])
df.info()
```

```python
# adding new column for weekday of schedule day and appointmnet

# weekdays of schedule day
df['schedule_weekday'] = df['schedule_day'].dt.strftime('%A')
# weekdays of schedule day
df['appointment_weekday'] = df['appointment_day'].dt.strftime('%A')
df.info()
```

```python
# converting no_show from string to integer for later analysis(like plotting)
df['no_show'] = df['no_show'].apply(lambda x: 0 if (x == 'No' or x == 'no') else 1)
```

```python
df.describe()
```

```python
df.query("handicap > 1")
```

```python
df.query("age < 0")
```

Only problem i can see is that **minimum age** is **-1** which is not possible and **handicap** values **greater than 1**. Maybe these are to show at what level a person is differently abled. Hence, I'll remove all rows with negitive ages and convert all handicap values greater than 1 to 1 as we are only interested at this point of time on whether a person is handicap or not.

```python
# converting all handicap values to 0 and 1
df['handicap'] = df['handicap'].apply(lambda x: 0 if x == 0 else 1)
```

```python
# removing all negative age values
df = df.query("age >= 0")
```

```python
df.describe()
```

```python
# set scholarship and sms as categorical variable

df['scholarship'] = pd.Categorical(df['scholarship'], categories = list(df['scholarship'].unique()), ordered = True)

df['sms_received'] = pd.Categorical(df['sms_received'], categories = list(df['sms_received'].unique()), ordered=True) 
```

```python
df.head()
```

```python
df.info()
```

```python
# save to a new file for future use 
df.to_csv('no_show_appointment_clean.csv', index = False)
```

## 4. Exploratory Data Analysis


**Question 1.** What is the proportion of people who show up for appointment v/s those who don't show up?

```python
# dataframe of all appointments where people showed up
df_show = df.query("no_show == 0")
# dataframe of all appointments where people didn't showed up
df_noshow = df.query("no_show == 1")
# calculating proportions
noshow_prop = df_noshow.shape[0] / df.shape[0]
show_prop = df_show.shape[0] / df.shape[0]
print('Proportion of appointments where patient didn\'t showed up:\t{:.4f}'.format(noshow_prop))
print('Proportion of appointments where patient showed up:\t\t{:.4f}\n'.format(show_prop))

# pie chart for same
plt.figure(figsize = [5,5])
ratios = [show_prop, noshow_prop]
labels = ['shows', 'no shows']
plt.pie(ratios, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in Appointments', fontweight = 'semibold', fontsize = 16);
```

> **Observation:**
- The number of times patients show up for appointment is greater than those who don't show up. Around 20.19% of appointments result in no show.

**Question 2.** Does Gender has any influence on patients showing up for appointment?  What are the proportion of males and females showing up and not showing up for appointments?

```python
# dataframe of all females
df_female = df.query("gender == 'F'")
# dataframe of all males
df_male = df.query("gender == 'M'")

# calculating proportions for females
noshow_prop = df_female.query("no_show == 1").shape[0] / df_female.shape[0]
show_prop = df_female.query("no_show == 0").shape[0] / df_female.shape[0]
ratio_female = [show_prop, noshow_prop]

# calculating proportions for males
noshow_prop = df_male.query("no_show == 1").shape[0] / df_male.shape[0]
show_prop = df_male.query("no_show == 0").shape[0] / df_male.shape[0]
ratio_male = [show_prop, noshow_prop]

plt.figure(figsize = [18, 8])

# plot pie chart for females
plt.subplot(1, 3, 1)
labels = ['shows', 'no shows']
plt.pie(ratio_female, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in Females', fontweight = 'semibold', fontsize = 12);    

# plot pie chart for males
plt.subplot(1, 3, 2)
labels = ['shows', 'no shows']
plt.pie(ratio_male, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in Males', fontweight = 'semibold', fontsize = 12);   

# plot barchart for both genders with total number of shows and no shows
plt.subplot(1, 3, 3)
g = sb.countplot(data = df, x = 'gender', hue = 'no_show');
plt.title('Number of Shows and No shows in Both Genders', fontweight = 'semibold', y = 1.1, fontsize = 12);
plt.xticks([0, 1], ['Female', 'Male']);
plt.xlabel('Gender');
plt.ylabel('no. of Patients');

plt.subplots_adjust(wspace = 0.5);
```

> **Observation:** 
- We can see that gender has almost no influence on whether a patient shows up for appoinment or not. In the last plot we can also see that number of female patients is greater than male patients.

**Question 3.** Whether receiving SMS Alerts for appointments have any affect on patients? Does it prompt them to show up for appointments?

```python
# dataframe of  all patients receiving sms
df_sms = df.query("sms_received == 1")
# dataframe of all patients not receiving sms
df_nosms = df.query("sms_received == 0")

# calculating proportions for sms
noshow_prop = df_sms.query("no_show == 1").shape[0] / df_sms.shape[0]
show_prop = df_sms.query("no_show == 0").shape[0] / df_sms.shape[0]
ratio_sms = [show_prop, noshow_prop]

# calculating proportions for nosmss
noshow_prop = df_nosms.query("no_show == 1").shape[0] / df_nosms.shape[0]
show_prop = df_nosms.query("no_show == 0").shape[0] / df_nosms.shape[0]
ratio_nosms = [show_prop, noshow_prop]

plt.figure(figsize = [18, 8])

# plot pie chart for sms
plt.subplot(1, 3, 1)
labels = ['shows', 'no shows']
plt.pie(ratio_sms, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in sms', fontweight = 'semibold', fontsize = 12);    

# plot pie chart for nosms
plt.subplot(1, 3, 2)
labels = ['shows', 'no shows']
plt.pie(ratio_nosms, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in No sms', fontweight = 'semibold', fontsize = 12);   

# plot barchart for both cases of sms with total number of shows and no shows
plt.subplot(1, 3, 3)
g = sb.countplot(data = df, x = 'sms_received', hue = 'no_show');
plt.title('Number of Shows and No shows in Both Genders', fontweight = 'semibold', y = 1.1, fontsize = 12);
plt.xticks([0, 1], ['No', 'Yes']);
plt.xlabel('sms received');
plt.ylabel('no. of appointments');

plt.subplots_adjust(wspace = 0.5);
```

> **Observation:**
- The proportion of patients who showed up but didn't receive SMS alerts is 83.3%, whereas the proportion of patients who showed and received SMS alerts is 72.43%.



**Question 4.** Does day of week on which appointment is fixed has any influence on patients showing up for appointment?  What are the proportion of people not showing up for appointments on different days of week?

```python
saturday = df.query('appointment_weekday == "Saturday"').shape[0] / df.shape[0]
saturday * 100
```

```python
def percent(plot, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.3f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 5
        y = p.get_y() + p.get_height() + 10
        ax.annotate(percentage, (x, y), size = 10)

        
base_color1 = sb.color_palette()[0]
base_color2 = sb.color_palette()[3]


plt.figure(figsize = [15, 5])

# plot for no of appointments on different days of week
plt.subplot(1, 2, 1)
cat_order2 = list(df['appointment_weekday'].value_counts().index)
ax = sb.countplot(data = df, x = 'appointment_weekday', order = cat_order2, color = base_color1);
percent(ax, df['appointment_weekday']);
plt.xticks(rotation = 40);
plt.title('Total Number of appointments on different days of week', y = 1.1, fontweight = 'semibold', fontsize = 12);  
plt.xlabel('Weekdays');
plt.ylabel('no. of appointments');

# plot for no. of appointments resulting in no show up on different days of week
plt.subplot(1, 2, 2)
df_noshow = df.query("no_show == 1")
cat_order1 = list(df_noshow['appointment_weekday'].value_counts().index)
ax = sb.countplot(data = df_noshow, x = 'appointment_weekday', order = cat_order1, color = base_color2);
percent(ax, df_noshow['appointment_weekday']);
plt.xticks(rotation = 40);
plt.title('Number of No-show appointments on different days of week', y = 1.1, fontweight = 'semibold', fontsize = 12); 
plt.xlabel('Weekdays');
plt.ylabel('no. of appointments');

plt.subplots_adjust(wspace = 0.6);
```

> **Observation:** 
- It can be clearly seen in first sub plot that number of appoinments on Saturday is very less. Also from second sublot we can infer that day of weekday has not much affect on patients showing up for appointments. 

**Question 5.** We would like to investigate whether scholarship has any influence on patients showing up for appointment? What are the proportion of no show appointments depending on whether they have scholarship or not? *Scholarship in dataset indicates whether or not the patient is enrolled in welfare program.* 

```python
# dataframe of all scholars
df_scholar = df.query("scholarship == 1")
# dataframe of all non scholars
df_no_scholar = df.query("scholarship == 0")

# calculating proportions for scholars
noshow_prop = df_scholar.query("no_show == 1").shape[0] / df_scholar.shape[0]
show_prop = df_scholar.query("no_show == 0").shape[0] / df_scholar.shape[0]
ratio_scholar = [show_prop, noshow_prop]

# calculating proportions for no_scholars
noshow_prop = df_no_scholar.query("no_show == 1").shape[0] / df_no_scholar.shape[0]
show_prop = df_no_scholar.query("no_show == 0").shape[0] / df_no_scholar.shape[0]
ratio_no_scholar = [show_prop, noshow_prop]

plt.figure(figsize = [18, 8])

# plot pie chart for scholars
plt.subplot(1, 3, 1)
labels = ['shows', 'no shows']
plt.pie(ratio_scholar, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in Scholars', fontweight = 'semibold', fontsize = 12);    

# plot pie chart for non scholars
plt.subplot(1, 3, 2)
labels = ['shows', 'no shows']
plt.pie(ratio_no_scholar, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False);
plt.title('Proportions of Shows and No shows in Non Scholars', fontweight = 'semibold', fontsize = 12);   

# plot barchart for both with total number of shows and no shows
plt.subplot(1, 3, 3)
g = sb.countplot(data = df, x = 'scholarship', hue = 'no_show');
plt.title('Number of Shows and No shows depending on Scholarship', fontweight = 'semibold', y = 1.1, fontsize = 12);
plt.xticks([0, 1], ['scholar', 'non scholar']);
plt.xlabel('Scholarship');
plt.ylabel('no. of appointments');

plt.subplots_adjust(wspace = 0.5);
```

> **Observation:** 
- Being enrolled in welfare program has postive affect on decreasing number of no show appointments. Percentage of no show appointments in case a patient is enrolled in welfare program is 23.74% whereas for patients not enrolled in welfare scheme the percentage is 19.81%. 

**6.** Do people with certain health conditions don't show up for appointment more often?

```python
# dataframe of all appointments where people showed up
df_show = df.query("no_show == 0")
# dataframe of all appointments where people didn't showed up
df_noshow = df.query("no_show == 1")

alch_show = df_show.query("alcoholism == 1").shape[0] / df_show.shape[0]
hypr_show = df_show.query("hypertension == 1").shape[0] / df_show.shape[0]
diab_show = df_show.query("diabetes == 1").shape[0] / df_show.shape[0]
hand_show = df_show.query("handicap == 1").shape[0] / df_show.shape[0]
other_show = df_show.query("handicap == 0 and alcoholism == 0 and hypertension == 0 and diabetes == 0").shape[0] / df_show.shape[0]
show_list = [alch_show, hypr_show, diab_show, hand_show, other_show]

alch_noshow = df_noshow.query("alcoholism == 1").shape[0] / df_noshow.shape[0]
hypr_noshow = df_noshow.query("hypertension == 1").shape[0] / df_noshow.shape[0]
diab_noshow = df_noshow.query("diabetes == 1").shape[0] / df_noshow.shape[0]
hand_noshow = df_noshow.query("handicap == 1").shape[0] / df_noshow.shape[0]
other_noshow = df_show.query("handicap == 0 and alcoholism == 0 and hypertension == 0 and diabetes == 0").shape[0] / df_show.shape[0]
noshow_list = [alch_noshow, hypr_noshow, diab_noshow, hand_noshow, other_noshow]


plt.figure(figsize = [18, 8])

# plot pie chart no show
plt.subplot(1, 3, 1)
labels = ['alcoholism', 'hypertension', 'diabetes', 'handicap', 'other']
plt.pie(noshow_list, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = True); # 
plt.title('Proportions of diseases in No Shows', fontweight = 'semibold', fontsize = 12);    

# plot pie chart for show
plt.subplot(1, 3, 2)
plt.pie(show_list, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = True);
plt.title('Proportions of diseases in Shows', fontweight = 'semibold', fontsize = 12); 

plt.subplots_adjust(wspace = 0.5);
```

> **Observation:** 
- We can see that patients suffering from alcoholism show up a little less than patients sufferong from other diseases. 

**7.** Do appointments in certain neighbourhood or locality have larger no show rates?

*Note*: as the number of neighbourhoods is large(around 81), I'll focus on 10 neighbourhoods with largest number of appointments.

```python
top10_neigh = list(df['neighbourhood'].value_counts().index)[:10]

plt.figure(figsize = [10, 5])
df_neigh = df[df['neighbourhood'].isin(top10_neigh)]

ct_count = df_neigh.groupby(['neighbourhood', 'no_show']).size()
ct_count = ct_count.reset_index(name = 'count')
ct_count
ax = sb.barplot(data=ct_count, x = 'neighbourhood', y = 'count' , hue = 'no_show', order = top10_neigh);
plt.xticks(rotation = 40);
plt.xlabel('Neighbourhood')
plt.ylabel('no. of appointments');
plt.title('Number of Shows and No-Shows depending on Neighbourhood', y = 1.1, fontsize = 14, fontweight = 'semibold');
```

> **Observation:** 
- Proportion of no show appointments in ITARARE neighbourhood is larger than other neighbourhoods where proportions of no show appoinments are almost same.

**Question 8.** Does patients in certain age bracket show up less or more for appointments?

```python
# to check for most common values for ride duration, so that outliers or extreme values are not plotted
df['age'].describe(percentiles = [0.99])
```

```python
plt.figure(figsize = [10, 7])

# dataframe of all appointments where people showed up
df_show = df.query("no_show == 0")
# dataframe of all appointments where people didn't showed up
df_noshow = df.query("no_show == 1")

bins = np.arange(0, 105, 5)
plt.hist(data = df_show, x = 'age', bins = bins, alpha = 0.6, label = 'Show');
plt.hist(data = df_noshow, x = 'age', bins = bins, alpha = 0.8, label = 'No Show');
plt.xlabel('age in years');
plt.ylabel('no. of appointments');
plt.title('Number of Show and No-Show appoinments for patients of Various Age', y = 1.0, fontweight = 'semibold', fontsize = 16);
plt.legend(loc = 'upper right');
```

> **Observation:** 
- There are no significant affect of age on patients showing for appointmnets.
